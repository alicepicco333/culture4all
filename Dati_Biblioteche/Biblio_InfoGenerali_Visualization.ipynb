```python
import pandas as pd
import json
from IPython.display import display

json_path=r"C:\Users\HWRUser\Desktop\DHDK Course\Information Visualization\Data_Italian_publishers\Dati_biblioteche\Biblio_Italia_General_info_2010_2011.json"
with open(json_path, 'r', encoding='utf-8') as file:
    json_data = json.load(file)


# Rimuovere i valori null
cleaned_data = {key: [entry for entry in value if entry is not None] for key, value in json_data.items()}

# Convertire in DataFrame
df_tav_4_3 = pd.DataFrame(cleaned_data["Tav 4.3"])
df_tav_4_3 = df_tav_4_3.replace('\n', '', regex=True)
df_tav_4_3= df_tav_4_3.fillna(value='')
df_tav_4_4 = pd.DataFrame(cleaned_data["Tav 4.4"])
df_tav_4_4= df_tav_4_4.fillna(value='')
df_tav_4_6 = pd.DataFrame(cleaned_data["Tav 4.6"])
df_tav_4_6 = df_tav_4_6.replace('\n', ' ', regex=True)
df_tav_4_6= df_tav_4_6.fillna(value='')

def allinea_testo(val):
    return 'text-align:  center'

# def allinea_testo_1(val):
#     return 'text-align:  left'

# Applica la formattazione alla riga 1
df_tav_4_3 = df_tav_4_3.style.applymap(lambda x: allinea_testo(x) if  x == 'ANNO 2010' or  x == 'ANNO 2011' else '')
df_tav_4_4 = df_tav_4_4.style.applymap(lambda x: allinea_testo(x) if  x == 'ANNO 2010' or  x == 'ANNO 2011' else '')
df_tav_4_6 = df_tav_4_6.style.applymap(lambda x: allinea_testo(x) if  x == 'ANNO 2010' or  x == 'ANNO 2011' else '')



display(df_tav_4_3)
display(df_tav_4_4)
display(df_tav_4_6)
```
