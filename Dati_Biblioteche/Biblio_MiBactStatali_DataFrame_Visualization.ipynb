```python
#Dataframes and JSON files of Biblioteche Mibact - Anno 2013

import pandas as pd
import json
from IPython.display import display

json_path=r"C:\Users\HWRUser\Desktop\DHDK Course\Information Visualization\Data_Italian_publishers\Dati_biblioteche\Biblio_Mibact_Statali_Ministero_2013.json"
with open(json_path, 'r', encoding='utf-8') as file:
    json_data = json.load(file)


# Rimuovere i valori null
cleaned_data = {key: [entry for entry in value if entry is not None] for key, value in json_data.items()}

# Convertire in DataFrame
df_tav_4_7 = pd.DataFrame(cleaned_data["Tav 4.7"])
df_tav_4_7 = df_tav_4_7.replace('\n', '', regex=True)
df_tav_4_7= df_tav_4_7.fillna(value='')
df_tav_4_8 = pd.DataFrame(cleaned_data["Tav 4.8"])
df_tav_4_8= df_tav_4_8.fillna(value='')
df_tav_4_9 = pd.DataFrame(cleaned_data["Tav 4.9"])
df_tav_4_9 = df_tav_4_9.replace('\n', ' ', regex=True)
df_tav_4_9= df_tav_4_9.fillna(value='')
df_tav_4_10 = pd.DataFrame(cleaned_data["Tav 4.10"])
df_tav_4_10 = df_tav_4_10.replace('\n', ' ', regex=True)
df_tav_4_10= df_tav_4_10.fillna(value='')
df_tav_4_10 = df_tav_4_10[['Column2', 'Column4', 'Column5', 'Column7']]
df_tav_4_10=df_tav_4_10.rename(columns={"Column2" : "Column3", 'Column7': 'Column6'})



#Rename columns
df_tav_1=df_tav_4_7.rename(columns={
    "Tavola 4.7 - Biblioteche, opere consultate, prestiti e personale delle biblioteche statali dipendenti dal Ministero dei beni e delle attività culturali e del turismo per provincia - Anno 2013 ": 'Tav. 1 - Numero di Biblioteche statali dipendenti dal MiBact per regioni e provincie, opere consultate e prestiti a privati e altre biblioteche - Anno 2013' 
})
df_tav_1 = df_tav_1.drop(columns=['Column6', "Column5"])
df_tav_1=df_tav_1.rename(columns={"Column3" : "Column2", "Column4": "Column7"})
df_tav_1=pd.concat([df_tav_1, df_tav_4_10], axis=1)
ordered_columns = ['Tav. 1 - Numero di Biblioteche statali dipendenti dal MiBact per regioni e provincie, opere consultate e prestiti a privati e altre biblioteche - Anno 2013',  'Column2', 'Column3', 'Column4', 'Column5', 'Column6', 'Column7']
df_tav_1 = df_tav_1.reindex(columns=ordered_columns)
display(df_tav_1)

df_tav_2=df_tav_4_8.rename(columns={"Tavola 4.8 - Consistenza del patrimonio librario delle biblioteche statali dipendenti dal Ministero dei beni e delle attività culturali e del turismo per tipologia e provincia - Anno 2013": 'Tav. 2 - Biblioteche statali dipendenti dal MiBact per tipologia e provincia  e consistenza del patrimonio librario - Anno 2013'})
display(df_tav_2)

df_tav_3 = df_tav_4_9.rename(columns={"Tavola 4.9 - Posti, lettori, lettori in totale e stranieri e opere consultate delle biblioteche statali dipendenti dal Ministero dei beni e delle attività culturali e del turismo per provincia e regione - Anno 2013": "Tav. 3 - Biblioteche statali dipendenti dal MiBact, posti per lettori, lettori in totale e stranieri - Anno 2013"})
df_tav_3 = df_tav_3.drop(columns=['Column6'])
display(df_tav_3)



json_path_tav_1 = r"C:\Users\HWRUser\Desktop\DHDK Course\Information Visualization\Data_Italian_publishers\Dati_biblioteche\Data_Biblioteche_Json\df_tav_1_prestiti_2013.json"
json_path_tav_2 = r"C:\Users\HWRUser\Desktop\DHDK Course\Information Visualization\Data_Italian_publishers\Dati_biblioteche\Data_Biblioteche_Json\df_tav_2_patrimonio_2013.json"
json_path_tav_3 = r"C:\Users\HWRUser\Desktop\DHDK Course\Information Visualization\Data_Italian_publishers\Dati_biblioteche\Data_Biblioteche_Json\df_tav_3_lettura_2013.json"



#Save df_tav_1 to JSON
df_tav_1.to_json(json_path_tav_1, orient='records', lines=True, force_ascii=False, indent=4)
# After exporting JSON
with open(json_path_tav_1, 'r') as f:
    data = f.read()

# Add square brackets
data = '[' + data.strip() + ']'

# Write back to the JSON file
with open(json_path_tav_1, 'w') as f:
    f.write(data)
    
with open(json_path_tav_1, 'r') as f:
    data = f.readlines()

# Join all lines to form a single string
json_content = ''.join(data)

# Split by '} ' to add commas
modified_json = json_content.replace('}\n', '},\n')  # Add comma after every '}'

# Write back to the JSON file
with open(json_path_tav_1, 'w') as f:
    f.write(modified_json)
    

# Save df_tav_2 to JSON
df_tav_2.to_json(json_path_tav_2, orient='records', lines=True, force_ascii=False, indent=4)
# After exporting JSON
with open(json_path_tav_2, 'r') as f:
    data = f.read()

# Add square brackets
data = '[' + data.strip() + ']'

# Write back to the JSON file
with open(json_path_tav_2, 'w') as f:
    f.write(data)
    
with open(json_path_tav_2, 'r') as f:
    data = f.readlines()

# Join all lines to form a single string
json_content = ''.join(data)

# Split by '} ' to add commas
modified_json = json_content.replace('}\n', '},\n')  # Add comma after every '}'

# Write back to the JSON file
with open(json_path_tav_2, 'w') as f:
    f.write(modified_json)
    

#Save df_tav_3 to JSON
df_tav_3.to_json(json_path_tav_3, orient='records', lines=True, force_ascii=False, indent=4)
# After exporting JSON
with open(json_path_tav_3, 'r') as f:
    data = f.read()

# Add square brackets
data = '[' + data.strip() + ']'

# Write back to the JSON file
with open(json_path_tav_3, 'w') as f:
    f.write(data)
    
with open(json_path_tav_3, 'r') as f:
    data = f.readlines()

# Join all lines to form a single string
json_content = ''.join(data)

# Split by '} ' to add commas
modified_json = json_content.replace('}\n', '},\n')  # Add comma after every '}'

# Write back to the JSON file
with open(json_path_tav_3, 'w') as f:
    f.write(modified_json)
```
